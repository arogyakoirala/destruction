```python
import geopandas as gpd
import pandas as pd
import numpy as np

shp = gpd.read_file(f"annots/Deir/Damage_Sites_Deir_ez_Zor_CDA.shp")
shp.head(6).T
```

```python
shp['SensDt2'] = shp['SensorDate']
```

```python
dcols = sorted([c for c in shp.columns if "SensDt" in c ])
icols = sorted([c for c in shp.columns if "DaSitCl" in c])
shp = shp[[*dcols, *icols, 'geometry']]
dcols, icols
```

```python
df = []
for k, row in shp.iterrows():
    entry = dict()
    for i in range(len(dcols)):
        if (row[dcols[i]] is not None):
            entry[row[dcols[i]]] = row[icols[i]]
    entry['geometry'] = row['geometry']
    df.append(entry)
newAnnots = gpd.GeoDataFrame(df)

mapping = {'Destroyed': 3, 'Severe Damage': 2, 'Moderate Damage': 1, 'No Visible Damage': 0, 'Impact Crater (Damage to Road)': np.nan, 'Impact Crater (Damage to Field)': np.nan }
for col in newAnnots.drop('geometry', axis=1).columns:
#     print(newAnnots[col].unique())
    newAnnots[col] = newAnnots[col].map(mapping)

newAnnots.to_file("deir-ez-zor_damage.gpkg")
```
